```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/sample_data/marketing_campaign.csv", sep='\t')
df.head()
```

```python
df.describe().T
```

```python
df.info()
```

```python
df[df["Income"].isnull()].index
```

```python
df["Income"] = df["Income"].fillna(df["Income"].mean())
```

```python
df[df["Income"].isnull()].index
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
numerical_columns=df.select_dtypes(include = 'number').columns
categorical_columns=df.select_dtypes(include = 'object').columns
```

```python
df[categorical_columns] =df[categorical_columns].astype('category')
categorical_columns
```

```python
df['Education'].unique()
```

```python
df['Marital_Status'].unique()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace(['Together', 'Married'] , 'Couple' )
df['Marital_Status'] = df['Marital_Status'].replace(['Divorced', 'Widow', 'Alone', 'Absurd', 'YOLO'] , 'Single')
```

```python
print("Total categories in the feature Marital_Status :\n", df["Marital_Status"].value_counts(), "\n")
print("Total categories in the feature Education :\n", df["Education"].value_counts())
```

```python
df.isna().sum()
```

```python
d_type = df.dtypes
n_unique = df.nunique()
null_ratio = df.isnull().sum() / df.shape[0]
pd.DataFrame({"D_Type" :  d_type , "N_Unique" : n_unique , "Null_Ratio" : null_ratio })
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y')
df.info()
```

```python
current_year = datetime.now().year
df['age'] = current_year - df['year_birth']
```

```python
df['total_amount_spent'] = df['mntwines'] + df['mntfruits'] + df['mntmeatproducts'] + df['mntfishproducts'] + df['mntsweetproducts'] + df['mntgoldprods']
df['total_amount_spent'].head(5)
```
